```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Given
C0 = 0 #kg/m3
C1 = 0.001 #kg/m3
C2 = 0.0012 #kg/m3
Q1 = 2000 #m3/s
Q2 = 1500 #m3/s
V = 1000000 #m3

a= (Q1*C1 + Q2*C2)
b= Q1 + Q2

#time array
sec = 60    #1 min = 60 sec
t = np.linspace(0, 30*sec, 500)

C= a/b - (a/b - C0)*(np.exp(-(b/V)* t))

#print(C)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t/60, C, label="Concentration in lake", color='blue')  # t/60 to convert seconds to minute
plt.axhline(a/b, color='red', linestyle='--', label=f"Steady-state = {a/b:.5f} kg/m³")
plt.xlabel('Time (minute)')
plt.ylabel('Concentration (kg/m³)')
plt.title('Lake Concentration vs Time')
plt.legend()
plt.grid(True)
plt.show()
```
